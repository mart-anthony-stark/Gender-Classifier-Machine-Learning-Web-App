## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ML Packages
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
#from sklearn.feature_extraction.text import TfidfVectorizer
```

## Load dataset

```python
df = pd.read_csv('name_gender_dataset.csv')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.dtypes
```

```python
sns.countplot(x=df['Gender'], data=df)
plt.show()
```

## Drop irrelevant columns

```python
df = df.drop(['Count', 'Probability'], axis=1)
df.head()
```

```python
df['Gender'].value_counts()
```

## Check null values

```python
df.isna().sum()
```

## Replace categorical sex values to numerical
- M: 0
- F: 1

```python
gender_map = {'M':0,'F':1}
df['Gender'].replace(gender_map, inplace=True)
df.head()
```

```python
df['Gender'].value_counts()
```

```python
X = df['Name']
```

## Feature extraction

```python
cv = CountVectorizer()
X = cv.fit_transform(X)
```

```python
cv.get_feature_names()
```

```python
X, y = X, df['Gender']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

## Training the model

### Printing predicted gender by names helper method



```python
def print_pred(names, predictions):
  for i in range(len(names)):
    if predictions[i] == 0:
      print(names[i], ": Male")
    else:
      print(names[i], ": Female")
```

# Naive Bayes MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
print("Accuracy: ",clf.score(X_train, y_train), "%")
```

### Multi-NB Prediction

```python
s_name = ['Mart Anthony', 'Andrea', 'Myka', 'Monica', 'Mark', 'John', 'Johhny', 'Marie']
vector = cv.transform(s_name).toarray()
```

```python
prediction = clf.predict(vector)
print_pred(s_name, prediction)
```
